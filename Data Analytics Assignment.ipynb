```python
#fill nan with median 
df['bmi'].fillna(df['bmi'].median(),inplace=True)
```

```python
df['bmi'].isnull().values.any()
```

```python
columns=('bmi','avg_glucose_level')
for col in columns:

    mean=df[col].mean()
    stdd=df[col].std()
    df=df[(df[col]<=mean+(3*stdd))]
df
```

```python
len(df)
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,MinMaxScaler

scaler = MinMaxScaler()

df['bmi'] = scaler.fit_transform(df[['bmi']])
df['age'] = scaler.fit_transform(df[['age']])
df['avg_glucose_level'] = scaler.fit_transform(df[['avg_glucose_level']])



df
```

```python
# unique values for each column
print(df['gender'].unique())
print(df['ever_married'].unique())
print(df['work_type'].unique())
print(df['Residence_type'].unique())
print(df['smoking_status'].unique())
```

```python
#labelencoder for 2 values
from sklearn.preprocessing import LabelEncoder
ever_married_encoder = LabelEncoder()
df['ever_married']=ever_married_encoder.fit_transform(df['ever_married'])
df
# print("Before Encoding:", list(df['ever_married'][-10:]))
# print("After Encoding:", ever_married_encoder_values[-10:])
```

```python
residence_type_encoder = LabelEncoder()
df["Residence_type"] = residence_type_encoder.fit_transform(df['Residence_type'])
df
#print("After Encoding:", residence_type_values[:5])
```

```python
#used onehotencoder because of ordinal attributes there are no continuos values 
from sklearn.preprocessing import OneHotEncoder
import numpy as np

gender_encoder = OneHotEncoder()
gender_reshaped = np.array(df['gender']).reshape(-1, 1) #reshape as function needs a 2-dimensional array as the input 
gender_values = gender_encoder.fit_transform(gender_reshaped)

smoke_status_encoder = OneHotEncoder()
smoke_status_reshaped = np.array(df['smoking_status']).reshape(-1, 1)
smoke_status_values = smoke_status_encoder.fit_transform(smoke_status_reshaped)

work_type_encoder = OneHotEncoder()
work_type_reshaped = np.array(df['work_type']).reshape(-1, 1)
work_type_values = work_type_encoder.fit_transform(work_type_reshaped)
```
