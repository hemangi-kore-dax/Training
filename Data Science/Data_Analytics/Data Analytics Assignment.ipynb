```python
import matplotlib.pyplot as plt
  
# frequencies

# setting the ranges and no. of intervals
range = (0, 100)
bins = 10  
  
# plotting a histogram
plt.hist(df.bmi, bins, range, color = 'green',
        histtype = 'bar', rwidth = 0.8)
  
# x-axis label
plt.xlabel('BMI')
# frequency label
plt.ylabel('Count')
# plot title
plt.title('My histogram')
  
# function to show the plot
plt.show()
```
